### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
# HINT use groupby method on Orbit column and get the mean of Class column
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from the given URL
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")

# Calculate the success rate for each orbit type
orbit_success_rate = df.groupby('Orbit')['Class'].mean() * 100

# Set the style for the plots (optional)
sns.set(style='whitegrid')

# Use bar plot to visualize the success rate of each orbit type
plt.figure(figsize=(10, 6))
sns.barplot(x=orbit_success_rate.index, y=orbit_success_rate.values, palette='viridis')

# Set the title and labels for the plot
plt.title('Success Rate of Each Orbit Type')
plt.xlabel('Orbit Type')
plt.ylabel('Success Rate (%)')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

Analyze the ploted bar chart try to find which orbits have high sucess rate.

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend. 


The function will help you get the year from the date:


```python
# A function to Extract years from the date 
year=[]
def Extract_year(date):
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
    
```

```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from the given URL
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")

# A function to Extract years from the date
def Extract_year(date):
    return date.split("-")[0]

# Add a new column 'Year' to the DataFrame with the extracted year from the Date
df['Year'] = df['Date'].apply(Extract_year)

# Convert 'Class' column to numeric (0: Failure, 1: Success)
df['Class'] = df['Class'].astype(int)

# Calculate the average success rate for each year
average_success_rate_by_year = df.groupby('Year')['Class'].mean() * 100

# Set the style for the plots (optional)
sns.set(style='whitegrid')

# Use line plot to visualize the average success rate yearly trend
plt.figure(figsize=(10, 6))
sns.lineplot(x=average_success_rate_by_year.index, y=average_success_rate_by_year.values, marker='o', color='b')

# Set the title and labels for the plot
plt.title('Average Launch Success Rate - Yearly Trend')
plt.xlabel('Year')
plt.ylabel('Average Success Rate (%)')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

you can observe that the sucess rate since 2013 kept increasing till 2020
